```python
import time
import ipywidgets as widgets
from IPython.display import display, clear_output

print("ü§ñ Smart Skincare AI - LLM Style System")

# UI Input
nama = widgets.Text(description="Nama:")
skin_type = widgets.Dropdown(
    options=["berminyak", "kering", "kombinasi", "sensitif"],
    description="Jenis kulit:"
)
problem = widgets.Dropdown(
    options=["jerawat", "kusam", "kering", "berminyak"],
    description="Masalah kulit:"
)
age = widgets.IntText(description="Umur:")
sensitif = widgets.Dropdown(
    options=["ya", "tidak"],
    description="Kulit sensitif?"
)

button = widgets.Button(description="üîç Analisis Kulit")
output = widgets.Output()

display(nama, skin_type, problem, age, sensitif, button, output)

# Fungsi AI
def analisis_ai(b):
    with output:
        clear_output()
        print("‚è≥ AI sedang menganalisis kondisi kulit kamu...")
        time.sleep(1)
        print("üîç Membandingkan data dengan knowledge base...")
        time.sleep(1)
        print("üß† Menghasilkan rekomendasi terbaik...\n")
        time.sleep(1)

        print(f"Hai {nama.value}! üíñ")

        if skin_type.value == "berminyak" and problem.value == "jerawat":
            print("\nüåû Morning Routine:")
            print("- Gentle facial wash")
            print("- Niacinamide serum")
            print("- Oil-free moisturizer")
            print("- Sunscreen SPF 50\n")

            print("üåô Night Routine:")
            print("- Facial wash")
            print("- Salicylic acid serum")
            print("- Centella serum")
            print("- Gel moisturizer")

        elif skin_type.value == "kering":
            print("\nüåû Morning Routine:")
            print("- Hydrating facial wash")
            print("- Hyaluronic acid serum")
            print("- Moisturizing cream")
            print("- Sunscreen")

        elif skin_type.value == "sensitif":
            print("\n‚ö†Ô∏è Routine untuk kulit sensitif:")
            print("- Low pH facial wash")
            print("- Centella asiatica serum")
            print("- Fragrance-free moisturizer")

        else:
            print("\n‚úÖ Basic Skincare Routine:")
            print("- Gentle facial wash")
            print("- Moisturizer")
            print("- Sunscreen")

button.on_click(analisis_ai)
```
